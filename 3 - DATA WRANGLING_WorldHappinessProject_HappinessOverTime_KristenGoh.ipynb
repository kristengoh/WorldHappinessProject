## Step 2 Merge Happiness Data Sets for All Years (6)

```python
dfs = [Happiness2015_v2, Happiness2016_v2,Happiness2017_v2, Happiness2018_v2, Happiness2019_v2, Happiness2020_v2 ]
HappinessAll = reduce(lambda left,right: pd.merge(left,right,on='Country', how='outer'), dfs).fillna('void')
```

```python
HappinessAll = pd.concat([Happiness2015_v2, Happiness2016_v2,Happiness2017_v2, Happiness2018_v2, Happiness2019_v2, Happiness2020_v2 ])
```
