## Step 3 Merge Health Data

```python
dfs = [LifeExpectancy, InfantMortality, SuicideRates, AlcoholSubstanceAbuse, MortalityRatePoisoning, Tobacco, DiseaseDeath]
```

```python
Health_All = reduce(lambda left,right: pd.merge(left,right,on='UID'), dfs)
```

```python
for col in Health_All.columns: 
    print(col) 
```

```python
Health_All2 = Health_All[['UID', 'Period', 'Dim1', 'Location', 'LifeExpectancyIndicator','LifeExpectancy', 'InfantMortalityIndicator', 'InfantMortality', 'SuicideRatesIndicator', 'SuicideRates', 'AlcoholSubstanceAbuseIndicator', 'AlcoholSubstanceAbuse', 'MortalityRatePoisoningIndicator', 'MortalityRatePoisoning', 'TobaccoIndicator', 'Tobacco', 'DiseaseDeathIndicator', 'DiseaseDeath']].copy()
```
